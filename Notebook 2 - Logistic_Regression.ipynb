```python
df = pd.read_csv('/Users/neelambabel/ADS Folder/Assignment 2/voice-data.csv')
# mark zero values as missing or NaN
df["meanfreq"] = df["meanfreq"].replace(0, np.NaN)
# fill missing values with mean column values
df.fillna(df.mean(), inplace=True)
# count the number of NaN values in each column
print(df.isnull().sum())
```

```python
df.isnull().values.any().sum()
```

```python
predictor_names = ['meanfreq',
 'sd',
 'median',
 'Q25',
 'Q75',
 'IQR',
 'skew',
 'kurt',
 'sp.ent',
 'sfm',
 'mode',
 'centroid',
 'meanfun',
 'minfun',
 'maxfun',
 'meandom',
 'mindom',
 'maxdom',
 'dfrange',
 'modindx']
```

```python
predictor_names
```

```python
target= ['label']
```

```python
target
```

```python
#setting predictor fields
x = df[predictor_names]
print(x.shape,'X')
#setting target
y = df[target]
print(y.shape,'Y')
```

```python
# setting Logistic regression classifier
# scikit-learn includes the intercept.

lr = LogisticRegression(C=1e9)

#dividing data to have a training and a testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= .2, random_state=0)

# Logistic regression metrics
LRm = lr.fit(x_train, y_train)



LRm.predict_proba(x_test)  # The returned estimates for all classes are ordered by the label of classes.
```

```python
LRm.coef_
```

```python
LRm.decision_function(x_test) # Predict confidence scores for samples.
```

```python
LRm.predict(x_test)
```

```python
y_pred = LRm.predict(x_test)
print("Accuracy score using Logistic regression is %s" %metrics.accuracy_score(y_test, y_pred))
```

```python
k=10
acc=[]
for i in range(k):
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state=i)
    # Logistic regression metrics
    LRm = lr.fit(x_train, y_train)
    a=metrics.accuracy_score(y_test, LRm.predict(x_test))
    acc.append(a)
    
print (acc)
print (np.mean(acc))
print (np.std(acc))
```

Summary of the stats:

* Is the relationship significant?
  Yes, the relationship is significant. We can see that the coefficient values,i.e. the slope, changes significantly,   implying the variations in the response variable (gender) w.r.t the predictor variables (meanfun, sp.ent etc) 

* Are any model assumptions violated? 
  The model assumes that the response variable should be categorical, hence there are no violations.

* Cross-validate the model. How well did it do?
  We split the training and test data into 80% and 20% resp. We cross-validated the consistency in accuracy score 10     times with different sets of training data and test data, and we've achieved an accuracy score of 97.2% with a         standard deviation of 4%.
